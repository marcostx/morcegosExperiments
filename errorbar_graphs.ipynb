# Estatísticas Gerais:  Accuracy, Precision, F1, Recall 

Neste notebook, são apresentados os gráficos de errorbar para as 4 métricas (accuracy, precision,f1 e recall). Os valores de cada gráfico representam a média obtida através de uma validação cruzada com k=10 folds, juntamente com o desvio padrão.

```python
import pandas as pd 
import numpy as np
import pandas
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
acc_cnn = [0.80000000000000004,0.73333333333333328,0.89473684210526316,
           0.6875,0.875,0.70588235294117652,0.90000000000000002,
          0.77272727272727271,0.86363636363636365,0.78260869565217395]
acc_mlp = [0.69565217391304346,0.80000000000000004,0.68181818181818177,
          0.59090909090909094,0.52631578947368418,0.6470588235294118,0.6875, 0.5625,
          0.66666666666666663,0.33333333333333331]
acc_dfa = [0.73913043478260865,0.45454545454545453,0.40909090909090912,
           0.69999999999999996,0.57894736842105265,0.52941176470588236,
           0.6875,0.8125,0.80000000000000004,0.73333333333333328]
plt.plot(acc_cnn,label="CNN")
plt.plot(acc_mlp,label='MLP')
plt.plot(acc_dfa,label='DFA')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# calculando as medias e desvios
mean_acc = np.mean(acc_cnn)
mean_mlp = np.mean(acc_mlp)
mean_dfa = np.mean(acc_dfa)
std_acc = np.std(acc_cnn)
std_mlp = np.std(acc_mlp)
std_dfa = np.std(acc_dfa)

mean_vals = [mean_acc,mean_mlp,mean_dfa]
std_vals = [std_acc,std_mlp,std_dfa]
print(mean_vals)
```

```python
titles = ['CNN', 'MLP', 'DFA']
width=0.35
y_pos  = np.arange(len(titles))

fig, ax = plt.subplots()
fig.set_size_inches(15.5, 10.5)
ax.bar(y_pos, mean_vals, width,yerr=std_vals, align='edge',
        color='red', ecolor='black')
ax.set_xticks(y_pos + width/2)
ax.set_xticklabels(('CNN', 'MLP', 'DFA'))
plt.ylim(0.0, 1.0)
ax.set_xlabel('Performance')
ax.set_title('Accuracy')
plt.show()
```
